```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("noise.jfif")
salt = cv2.imread("salt.jfif")
```

```python
plt.imshow(image)
```

### Mean Fliter

```python
average = cv2.blur(image,(5,5))
```

```python
plt.imshow(average)
```

### Gaussian Filter

```python
gblur = cv2.GaussianBlur(image,(5,5),0)
```

```python
plt.imshow(gblur)
```

### Median Filter

```python
median = cv2.medianBlur(salt,5)
```

```python
plt.imshow(median)
```

### Bilateral Filter

```python
bfilter = cv2.bilateralFilter(image , 9 , 75 ,75)
```

```python
plt.imshow(bfilter)
```
