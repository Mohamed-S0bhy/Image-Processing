```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("sodoku.jfif")
```

```python
plt.imshow(image)
```

### Laplacian

```python
lab = cv2.Laplacian(image,cv2.CV_64F,ksize = 3)
lab = np.uint8(np.absolute(lab))
plt.imshow(lab)
```

### Solbel 

```python
sobely = cv2.Sobel(image,cv2.CV_64F,1,0)
sobely = np.uint8(np.absolute(sobely))
plt.imshow(sobely)
```

```python
sobelx = cv2.Sobel(image,cv2.CV_64F,0,1)
sobelx = np.uint8(np.absolute(sobelx))
plt.imshow(sobelx)
```

```python
sobelxy = cv2.bitwise_or(sobelx,sobely)
plt.imshow(sobelxy)
```
