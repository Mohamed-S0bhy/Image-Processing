```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = np.zeros((300,300,3),np.uint8)
img1 = cv2.rectangle(img1 , (0,0),(150,300),(255,255,255),-1)
```

```python
plt.imshow(img1)
```

```python
img2 = np.zeros((300,300,3),np.uint8)
img2 = cv2.rectangle(img2 , (50,0),(250,150),(255,255,255),-1)
```

```python
plt.imshow(img2)
```

```python
And = cv2.bitwise_and(img1,img2)
plt.imshow(And)
```

```python
Or = cv2.bitwise_or(img1,img2)
plt.imshow(Or)
```
